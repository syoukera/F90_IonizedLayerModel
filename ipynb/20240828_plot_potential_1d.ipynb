```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
list_step = np.linspace(10000, 100000, 10)
# list_step = np.linspace(100000, 1000000, 10)
# list_step = [1, 101, 201]

dfs = []

for step in list_step:

    # path = f'..\output\omega_V2.0_omega_ion0.005\potential_1d_{step:.0f}.dat'
    path = f'..\potential_1d_{step:.0f}.dat'
    print(path, os.path.exists(path))

    df = pd.read_csv(path, sep='\s+')
    dfs.append(df)
```

```python
df.columns
```

```python
for columns in df.columns[1:-1]:

    df.plot(x='X[m]', y=columns)

    plt.ylabel(columns)
```

```python
for columns in df.columns[1:-1]:

    plt.figure()
    ax = plt.gca()

    for i, df in enumerate(dfs[:]):

        df.plot(x='X[m]', y=columns, ax=ax, label=f'{list_step[i]:.0f}')

    plt.ylabel(columns)
```
